## ***3. Pre-processing : Nettoyage, transformation et enrichissement des DataFrames***

### ***(a) Gestion des valeurs manquantes dans df_geo_names***

```python
# On supprime les colonnes avec des valeurs manquantes
colonnes_a_supprimer = ['latitude', 'longitude', 'éloignement']

# Supprimer les colonnes spécifiées
df_geo_names = df_geo_names.drop(columns=colonnes_a_supprimer)

# Apperçu du df et du shape
display(df_geo_names.head())
print('----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
display(df_geo_names.shape)
print('----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
# Vérification des valeurs manquantes
print(df_geo_names.isna().sum())
```

### ***(b) Création de 4 nouvelles catégories de taille d'entreprises sur df_etab_effectifs (selon le format de l'INSEE) et suppression des autres***

```python
# Nous créons 4 nouvelles catégories de taille d'entreprises, selon le format de l'INSEE
    # micro entreprise    : de 0 à 9 salariés
    # petite entreprise   : de 10 à 49 salariés
    # moyenne entreprise  : de 50 à 199 salariés
    # grande entreprise   : plus de 200 salariés

# Création des colonnes micro_E, petites_E, moyennes_E, grandes_E
df_etab_effectifs['micro_E'] = df_etab_effectifs[['E14TS0ND', 'E14TS1', 'E14TS6']].astype(int).sum(axis=1)
df_etab_effectifs['petites_E'] = df_etab_effectifs[['E14TS10', 'E14TS20']].astype(int).sum(axis=1)
df_etab_effectifs['moyennes_E'] = df_etab_effectifs[['E14TS50', 'E14TS100']].astype(int).sum(axis=1)
df_etab_effectifs['grandes_E'] = df_etab_effectifs[['E14TS200', 'E14TS500']].astype(int).sum(axis=1)
df_etab_effectifs['total_entreprises'] = df_etab_effectifs[['micro_E', 'petites_E', 'moyennes_E', 'grandes_E']].sum(axis=1)

# Vérif du df et du shape
display(df_etab_effectifs.head())
print('----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
display(df_etab_effectifs.shape)
```

```python
# Colonnes à supprimer
colonnes_a_supprimer = [
    'E14TS0ND', 'E14TS1', 'E14TS6', 'E14TS10',
    'E14TS20', 'E14TS50', 'E14TS100', 'E14TS200', 'E14TS500']

# Supprimer les colonnes spécifiées
df_etab_effectifs = df_etab_effectifs.drop(columns=colonnes_a_supprimer)

# Apperçu du df et du shape
display(df_etab_effectifs.head())
print('----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
display(df_etab_effectifs.shape)
```

### ***(c) Création d'une correspondance entre tous les df au niveau du code géographique (CODGEO et code_Insee)***

#### ***- Traitement du df_geo_names***

```python
# Rajout d'un zéro dans la variable code_insee pour la transformer en CODGEO pour préparer la fusion des fichiers

# Fonction pour ajouter un zéro au début
def add_zero_start(df, column):
    df[column] = df[column].apply(lambda x: '0' + str(x))
    return df

# Utilisation de la fonction pour transformer 'code_insee'
df_geo_names = add_zero_start(df_geo_names, 'code_insee')

# Renommer la colonne 'code_insee' en 'CODGEO'
df_geo_names = df_geo_names.rename(columns={'code_insee': 'CODGEO'})

# Verif du df et du shape
display(df_geo_names.head())
print('----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
display(df_geo_names.shape)
```

### ***(c) Comparaison du salaire moyen par heure par catégorie d'emploi dans les principales villes***

```python
# Liste des principales villes
principales_villes = ['Paris', 'Marseille', 'Lyon', 'Toulouse', 'Nice',
                      'Nantes', 'Strasbourg', 'Montpellier', 'Bordeaux', 'Lille']

# Filtrer les données pour inclure seulement les principales villes
data_filtered = df_salary_town[df_salary_town['LIBGEO'].isin(principales_villes)]
data_filtered = data_filtered[['LIBGEO', 'SNHMC14', 'SNHMP14', 'SNHME14', 'SNHMO14']]

# Reformer les données pour qu'elles soient adaptées à un graphique seaborn
data_melted = data_filtered.melt(id_vars='LIBGEO', value_vars=['SNHMC14', 'SNHMP14', 'SNHME14', 'SNHMO14'],
                                 var_name='Categorie', value_name='Salaire')

# Renommer les valeurs de catégories pour mieux les comprendre
categorie_map = {
    'SNHMC14': 'Cadres',
    'SNHMP14': 'Profession intermédiaire',
    'SNHME14': 'Employés',
    'SNHMO14': 'Ouvriers'
}
data_melted['Categorie'] = data_melted['Categorie'].map(categorie_map)

# Créer un graphique à barres groupées
plt.figure(figsize=(10, 8))
sns.barplot(data=data_melted, x='LIBGEO', y='Salaire', hue='Categorie')
plt.title('Comparaison du salaire moyen par heure par catégorie d\'emploi dans les principales villes')
plt.xlabel('Ville')
plt.ylabel('Salaire moyen par heure (Euros)')
plt.xticks(rotation=45)
plt.legend(title='Catégorie')

# Afficher le graphique
plt.tight_layout()
plt.show()
```

***Conclusion***

* Sur le salaire moyen horaire des cadres : il varie entre 22 et 33 euros de l'heure
 * On constate une nette disparité entre le salaire moyen horaire des cadres et ceux des autres catégories professionnelles
 * De la même façon ce salaire moyen horaire des cadres varient fortement d'une ville comme Paris par rapport à Montpellier
* Sur le salaire moyen horaire des professions intermédiaires: il varie entre 12 et 16 euros de l'heure
  * Les écarts sur cette  catégorie sont moindres
* Sur le salaire moyen horaire des employés et ouvriers : il se situe entre 10 et 14 euros de l'heure


### ***(d) Répartition des modes de cohabitation par ville***

### ***(e) Test anova entre les régions du territoire***

```python
# Test anova entre les régions du territoire
# Hypothèse :
# H0 : Aucune influence significative entre les régions et le salaire net moyen
# H1 : Influence significative entre les régions et le salaire net moyen
import statsmodels.api as sm
from statsmodels.formula.api import ols

result = ols('SNHM14 ~ C(REG)', data=df_final).fit()
anova_table = sm.stats.anova_lm(result, typ=2)

# Affichage de la table ANOVA
print(anova_table)

# Interprétation des résultats
alpha = 0.05
if anova_table["PR(>F)"][0] < alpha:
    print("p-value (PR(>F)) < 5%  - Nous rejetons H0 et acceptons H1 : Il y a une influence significative entre les régions et le salaire net moyen.")
else:
    print("p-value (PR(>F)) >= 5%  - Nous ne rejetons pas H0 : Il n'y a pas d'influence significative entre les régions et le salaire net moyen.")
```

***Conclusion***
* Ces résultats ANOVA montrent que le test est extrêmement significatif. Le F-score élevé (82) et la p-value de 0 indiquent qu'il y a une influence significative des régions (REG) sur le salaire net moyen (SNHM14).

* En d'autres termes, nous rejetons l'hypothèse nulle (H0) et acceptons l'hypothèse alternative (H1)

```python
# Début du ML avec l'entraînement des différents modèles
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
import xgboost as xgb
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Sélectionner les variables pertinentes (d'après la matrice de corrélation)
X = df_final[['REG', 'Ecart_Salarial_HF', 'densite_diplomes_BAC5', 'Taux_Emploi', 'pourcentage_plus_50','salaire_hommes', 'salaire_femmes', 'salaire_cadres', 'salaire_employes', 'salaire_ouvriers', 'salaire_plus_50']]
y = df_final['salaire_moyen']

# Encodage des colonnes non numériques
X = pd.get_dummies(X, drop_first=True)

# Standardisation des données
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Séparation en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Modèle de régression linéaire
model_linear = LinearRegression()
model_linear.fit(X_train, y_train)

# Modèle XGBoost
model_xgb = xgb.XGBRegressor()
model_xgb.fit(X_train, y_train)

# Modèle Random Forest
model_rf = RandomForestRegressor(random_state=42)
model_rf.fit(X_train, y_train)

# Modèle K-Nearest Neighbors
model_knn = KNeighborsRegressor()
model_knn.fit(X_train, y_train)

# Prédictions
y_pred_linear = model_linear.predict(X_train)
y_pred_xgb = model_xgb.predict(X_train)
y_pred_rf = model_rf.predict(X_train)
y_pred_knn = model_knn.predict(X_train)

# Évaluation des modèles
def evaluate_model(y_true, y_pred, model_name):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    print(f"\nÉvaluation du modèle {model_name}:")
    print('Mean Squared Error:', mse)
    print('Root Mean Squared Error:', rmse)
    print('Mean Absolute Error:', mae)
    print('R-squared:', r2)

# Évaluation des quatre modèles
evaluate_model(y_train, y_pred_linear, "Régression Linéaire")
evaluate_model(y_train, y_pred_xgb, "XGBoost")
evaluate_model(y_train, y_pred_rf, "Random Forest")
evaluate_model(y_train, y_pred_knn, "K-Nearest Neighbors")
```
